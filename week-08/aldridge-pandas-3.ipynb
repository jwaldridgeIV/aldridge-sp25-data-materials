Just isolate people who work on a farm in some capacity. Create a filter and subset the data using str.contains to return rows whose occupation contains FARM. How many people work on a farm?

```python
farm_filter = utah_df['occupation'].str.contains('FARM')
how_many_farmers = utah_df[farm_filter].count()
farmer_total = how_many_farmers['occupation']

print(f'There were {farmer_total} people working on or around farms in Utah during this time period!')
```

The three most common categories for marrystatus have been shortened to single letters. Write three lines of code that each uses str.replace on the marrystatus column to replace:
m with married
s with single
w with widowed

```python
def replace_marrystatus(some_string):
    if some_string == "m":
        return "married"
    elif some_string == "s":
        return "single"
    elif some_string == "w":
        return "widowed"
```

```python
utah_df['marrystatus'] = utah_df["marrystatus"].apply(replace_marrystatus)
```

Define a function called birthyear_calc to calculate the year an individual was born based on their age column and the year the census was recorded (1880) - note: this won’t be perfect.

```python
def birthyear_calc(current_age):
    birthyear = 1880 - current_age
    return birthyear

birthyear_calc(40)
```
